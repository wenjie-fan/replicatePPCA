```python
# Load target data
raw = pd.read_csv('../datasets/pSCoPE/raw.txt', sep='\t', index_col=0).T
raw = raw.drop('peptide', axis=0)

# Combining repetitive proteins in the target raw data
unique_proteins = raw.columns.unique()
result = pd.DataFrame()
for protein in unique_proteins:
    sub_raw = raw.loc[:, raw.columns==protein] # All columns of the protein
    protein_mean = sub_raw.mean(axis=1, skipna=True) # Mean skipping NaNs
    result[protein] = protein_mean # Record to result dataframe
data = result.sort_index(axis=1) # Sort protein names
```

```python
# Load meta data
guide = pd.read_csv('../datasets/pSCoPE/guide.csv',index_col=0)
guide = guide[guide.index.isin(data.index)]

# Create format required for emPPCA
np.random.seed(123)
R = 40 
id_rep = [] 
label_rep = [] 
Y = [] 
for r in range(R):
    id_rep.append(guide[guide['Raw.file'] == f'XwGH05{r:02d}'].index)
    label_rep.append(guide.loc[id_rep[r]]['celltype'].tolist())
    Y.append(data.loc[id_rep[r]])
    
# Concatenate replicates to standardise together
Y_concat = np.concatenate(Y)
Y_concat = StandardScaler().fit_transform(Y_concat)

# Split replicates back to the required form
counter = 0
for r in range(R):
    Y[r] = Y_concat[counter:(counter+Y[r].shape[0]), :]
    counter += Y[r].shape[0]

# Text labels and color labels
label = [label_rep[i] for i in range(R)]
color_dict = {'untreated': 0, 'LPS': 1}
col = [[color_dict[i] for i in label[r]] for r in range(R)] 
```
